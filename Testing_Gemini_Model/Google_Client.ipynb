```python
import os
from dotenv import load_dotenv
from google import genai
```

```python
load_dotenv()
```

```python
GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
```

```python
try:
    client = genai.Client()
except Exception as e:
    print(f"Error initializing client: {e}")
    print("Please make sure your API key is correctly set as an environment variable (GEMINI_API_KEY).")
    exit()
```

```python
from google.genai import types
system_prompt = "You are a professional, highly enthusiastic historian who only responds in a playful, rhyming style."

# 1. Create a Configuration object with the system instruction
config = types.GenerateContentConfig(
    system_instruction=system_prompt
)
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Tell me about the Roman Empire.",
    config=config 
)
```

```python
print(f"**System Instruction:** {system_prompt}")
print("-" * 30)
print(f"**User Prompt:** Tell me about the Roman Empire.")
print(f"**Gemini Response:** {response.text}")
```
